```python
import requests
from bs4 import BeautifulSoup

def scrape_amazon_mobiles():
    url = 'https://www.amazon.in/s?k=mobile+phones&rh=n%3A1389401031'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    products = soup.find_all('div', class_='s-result-item')
    scraped_data = []

    for product in products:
        title = product.find('h2').find('a').find('span').text.strip()
        price = product.find('div').find('span').find('div').find('a').find('span').find('span').text.strip()
        url = product.find('h2').find('a')['href']

        item = {
            'title': title,
            'price': price,
            'url': url
        }

        scraped_data.append(item)

    return scraped_data


scraped_data = scrape_amazon_mobiles()
scraped_data[:5]
```

```python
import pandas as pd
df = pd.DataFrame(scraped_data)
df.head()
```

```python
df.to_csv('amazon_mobiles_data.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_amazon_mobiles(pages=3):
    base_url = 'https://www.amazon.in/s?k=mobile+phones&rh=n%3A1389401031'
    scraped_data = []

    for page in range(1, pages + 1):
        url = f'{base_url}&page={page}'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        products = soup.find_all('div', class_='s-result-item')

        for product in products:
            title_elem = product.find('h2')
            title = title_elem.find('a').find('span').text.strip() if title_elem and title_elem.find('a') else 'Not Available'

            price_elem = product.find('div').find('span').find('div').find('a').find('span').find('span')
            price = price_elem.text.strip() if price_elem else 'Not Available'

            url_elem = product.find('h2').find('a')
            url = url_elem['href'] if url_elem else 'Not Available'

            item = {
                'title': title,
                'price': price,
                'url': url
            }

            scraped_data.append(item)

    return scraped_data
scraped_data = scrape_amazon_mobiles(pages=3)
```

```python
import pandas as pd
df = pd.DataFrame(scraped_data)
df.to_csv('/content/amazon_mobiles_data.csv.zip', index=False)
```
